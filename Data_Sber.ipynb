```python
K = 50
y_predict = []
for id_ in id_predict:
    nuni = pd_all[pd_all['user_id'] == id_].product_id.nunique()
    if (nuni >= K):
        y_predict.append(pd_all[pd_all['user_id'] == id_].product_id.unique()[-K:])
    else:
        print("small future ", id_, nuni)
        y_predict.append(np.concatenate((pd_all[pd_all['user_id'] == id_].product_id.unique(), top50popular[:K-nuni])))
```
